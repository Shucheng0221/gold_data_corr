```python
from sklearn.linear_model import ElasticNet
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler, MinMaxScaler

# 按长度滚动
def length_rolling_training(data, features, label, window_size=500, train_size=0.8, step=1):
    results = []

    # 获取数据的总长度
    total_length = len(data)

    # 计算每个滚动窗口的起始点
    start_indices = range(0, total_length - window_size, step)

    for start_idx in start_indices:
        # 定义当前窗口的起始和结束索引
        end_idx = start_idx + window_size

        # 划分训练集和测试集
        train_data = data.iloc[start_idx:int(start_idx + window_size * train_size)]
        test_data = data.iloc[int(start_idx + window_size * train_size):end_idx]

        # 检查训练集和测试集是否为空或样本数不足
        if train_data.empty or test_data.empty:
            print(f"Skipping window: {start_idx} to {end_idx} (insufficient data)")
            continue

        # 提取特征和标签
        X_train, y_train = train_data[features], train_data[label]
        X_test, y_test = test_data[features], test_data[label]
        
        scaler = StandardScaler()
        X_train = scaler.fit_transform(X_train)
        X_test = scaler.transform(X_test)
        
        # 降维
        pca = PCA(n_components=0.95)  
        # X_train = pca.fit_transform(X_train)
        # X_test = pca.transform(X_test)

        # 再次检查特征和标签是否为空或样本数不足
        if len(X_train) == 0 or len(X_test) == 0:
            print(f"Skipping window: {start_idx} to {end_idx} (empty X or y)")
            continue

        # 初始化并训练模型
        model = LinearRegression()
        model = ElasticNet(alpha=0.3, l1_ratio=0)
        model.fit(X_train, y_train)

        # 预测
        y_test_pred = model.predict(X_test)

        # 存储预测结果
        test_result = pd.DataFrame(
            {
                label: y_test,
                'pre': y_test_pred
            },
            index=y_test.index,
        )
        results.append(test_result)

    # 合并所有结果
    if results:
        final_result = pd.concat(results)
        return final_result
    else:
        print("No valid results generated.")
        return None

window_size=500
train_size=0.8
features = daily_data.columns.drop('gold_r_future').tolist()
target = 'gold_r_future'
result = length_rolling_training(
                                data=daily_data.dropna().replace([-np.inf,np.inf],0),
                                features=features,  # 特征列
                                label=target,  # 标签列
                                window_size=window_size, 
                                train_size=train_size, 
                                step=int(window_size*(1-train_size))
                                )

result
```

```python
def r2(y_pred,y):
    # 计算 R²
    #ss_total = np.sum((y - np.mean(y)) ** 2)  # 总平方和
    ss_residual = np.sum((y - y_pred) ** 2)   # 残差平方和

    # r_squared = 1 - (ss_residual / ss_total)
    # print(f"R²: {r_squared}")
    return ss_residual

mse=r2(y_pred=result['pre'],y=result[target])
mse
```

一级指标的滚动贝塔和特质波动率

```python
import statsmodels.api as sm

def roll_regress_feature(data_monthly1,feature,target,window_size=36):
    data_monthly = data_monthly1.replace([np.inf,-np.inf],np.nan)
    data_monthly = data_monthly.dropna()
    # 窗口长度
    window_size = 36
    # 初始化结果列表
    results = []
    # 滚动窗口回归
    for start in range(len(data_monthly) - window_size + 1):
        # 定义窗口数据
        window = data_monthly.iloc[start:start + window_size]
        
        # 自变量和因变量
        X = window[feature]
        y = window[target]
        
        # 添加截距项
        X = sm.add_constant(X)
        
        # 回归模型
        model = sm.OLS(y, X).fit()
        
        # 预测值和残差
        y_pred = model.predict(X)
        residuals = y - y_pred
        
        # 记录贝塔值和时间戳
        beta = model.params
        timestamp = window.index[-1]  # 窗口的最后一个时间戳
        residual_list = np.array(residuals.tolist())  # 转换为列表存储
        
        # 存储结果
        results.append({
            'date': timestamp,
            **beta.to_dict(),  # 将系数字典展开为列
            'IV': np.std(residual_list)  # 存储特质波动率
        })

    # 转换为 DataFrame
    results_df = pd.DataFrame(results)
    return results_df.set_index('date')
```

日度指标

```python
results_df = pd.DataFrame()

for col in daily_data.columns.drop('gold_r_future'):
# 查看结果
    results_df_single = roll_regress_feature(daily_data,[col],'gold_r_future',360)
    results_df_single = results_df_single.rename(columns={'const':'const_'+col,'IV':'IV_'+col})
    results_df=pd.concat([results_df,results_df_single],axis=1)
results_df
```
