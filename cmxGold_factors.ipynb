一级指标的滚动贝塔和特质波动率

```python
import statsmodels.api as sm

def roll_regress_feature(data_monthly1,feature,target,window_size=36):
    data_monthly = data_monthly1.replace([np.inf,-np.inf],np.nan)
    data_monthly = data_monthly.dropna()
    # 窗口长度
    window_size = 36
    # 初始化结果列表
    results = []
    # 滚动窗口回归
    for start in range(len(data_monthly) - window_size + 1):
        # 定义窗口数据
        window = data_monthly.iloc[start:start + window_size]
        
        # 自变量和因变量
        X = window[feature]
        y = window[target]
        
        # 添加截距项
        X = sm.add_constant(X)
        
        # 回归模型
        model = sm.OLS(y, X).fit()
        
        # 预测值和残差
        y_pred = model.predict(X)
        residuals = y - y_pred
        
        # 记录贝塔值和时间戳
        beta = model.params
        timestamp = window.index[-1]  # 窗口的最后一个时间戳
        residual_list = np.array(residuals.tolist())  # 转换为列表存储
        
        # 存储结果
        results.append({
            'date': timestamp,
            **beta.to_dict(),  # 将系数字典展开为列
            'IV': np.std(residual_list)  # 存储特质波动率
        })

    # 转换为 DataFrame
    results_df = pd.DataFrame(results)
    return results_df.set_index('date')
```

```python
results_df_day = pd.DataFrame()

# 创建图形和主轴
fig, ax1 = plt.subplots(figsize=(10, 6))

# 循环处理每个列并绘制 IV
for col in daily_data.drop(columns=[target, 'GC01.CMX收盘价']).columns:
    results_df_single = roll_regress_feature(daily_data, [col], target, 360)
    results_df_single = results_df_single.rename(columns={'const': 'const_' + col, 'IV': 'IV_' + col})
    
    # 绘制 IV 曲线
    ax1.plot(results_df_single['IV_' + col], label='IV_' + col)
    #ax1.plot(results_df_single[col], label= col)
    results_df_day = pd.concat([results_df_day, results_df_single], axis=1)

# 为主轴添加图例
ax1.legend(loc='upper left', bbox_to_anchor=(1.05, 1))
ax1.set_ylabel("IV")

# 创建副轴
ax2 = ax1.twinx()
ax2.plot(daily_data['GC01.CMX收盘价'], color='black', linestyle='--', label='gold_close')
ax2.set_ylabel("Gold Close Price")

# 为副轴添加图例
ax2.legend(loc='upper right', bbox_to_anchor=(1.3, 0.9))

# 调整布局并显示
plt.tight_layout(rect=[0, 0, 1.3, 1])
plt.rcParams['font.sans-serif'] = ['SimHei'] 
plt.rcParams['axes.unicode_minus'] = False 
plt.show()
```
